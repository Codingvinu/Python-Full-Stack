```python
#Bitwise Operator
"""
These are used for bit level operations
these are the symbols
&- Bitwise and
|- Bitwise or
^ - Bitwise xor
~ -Bitwise not
>>-right shift
<<- left shift
"""

```

```python
"""1. Bitwise and
It works simlilar to logical and 
it compares bits at positions in both the operands and decides what should be the bit at position in the resultant 
if both the bits are 1 then the resultant will be 1, else 0
"""

a = 5
b =3
print(a & b)
```

```python
""" Bitwise or
in the resultant , the bit at any specific position will be 1 
if any of the two bits under consideration are 1 else 0
"""
a=5
b=3
print(a|b)
```

```python
""" Bitwise XOR
if both the bits are different , then it will be 1 else 0
"""
a=5
b=3
print(a^b)
```

```python
"""Bitwise not
this is an unary operator
it inverses the bits present in given number
~n=-n-1
"""
a=5
print(~5)#-5-1=-6
b=-5
print(~-5)#-(-5)-1=4
```

```python
"""5. left shift(<<)
this operator is used to move all the bits to left by given number of position
the result of a << n will be a*2^n
"""
a=50
print(a<<5)#a* 2^n
```

```python
"""6. Right shift(>>)
this is operator is used to move all the bits to right by given number of positions 
the result of a>> n will be a/2^n"""
a=100
print(a>>2)
```
