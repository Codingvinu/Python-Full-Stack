```python
#Data Types 
"""Data types define the type of data a variable can store"""
#Numeric Category
#integer(integer data type is used to store the integer data)
a=20;
print(a);
type(a)
```

```python
#Float(float data type is used to store floatting values)
a=30.2
print(a)
type(a)
```

```python
#Complex(complex data type means combination of integer and floating data )
z = 2 + 3j
print(z)
type(z)
```

```python
#ordered Data types[group of elements in same order]
#String(string data types is used to store charecters)
a="vinod"
print(a)
type(a)
```

```python
#list[]
'''
A list is a collection of items that is:
Ordered
Mutable (can be changed)
Allows duplicate values
Written using square brackets [ ]
'''
a=[1,2,3]
print(a)
type(a)
```

```python
#tuple()
'''A tuple is similar to a list but:
Ordered
Immutable (cannot be changed)
Allows duplicates
Written using parentheses ( )'''
a=(20,30)
print(a)
type(a)
```

```python
#Range The range() function is used to generate a sequence of numbers.
for i in range(1, 5):
    print(i)
```

```python
#unordered data types{group of elements different order}
#set
a={1,4,6,6,7}
print(a)
type(a)
```

```python
#mapping data types
# Dictionary( A dictionary stores data in key–value pairs)
a={'a':1,'b':2}
print(a)
type(a)
```

```python
#boolean data type(true /false)
a=10;
b=20;
print(a>b)
print(b>a)
```

```python
#variable
# A variable is used to store data that can be used and changed later in a program.
a=20;
b="vinod";
print(a,b)
```

```python
#Type --type() is a built-in function used to find the data type of a value or variable.
a=20
type(a)
```

```python
#swapping
a=10;
b=20;
c=a;
a=b;
b=c;
print(a,b)
```

```python
#delete 
a=10;
del(a)
```
